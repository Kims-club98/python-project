```python
import matplotlib as mpl
import matplotlib.pyplot as plt

#한글설정
plt.rc('font',family='Malgun Gothic')
#사이즈값 설정
plt.rc('font',size=20)
#한글깨지 방지
plt.rc('axes',unicode_minus=False) #xy축 -축의 한글 깨짐 방지(영문 사용 시 필요 없음)
```

```python
x=[1,2,3,4,5]
y=[3,6,9,12,15]
```

```python
plt.plot(x,y)
plt.xticks([x for x in range(min(x),max(x)+1,1)],) # 소괄호 내 대괄호(list임)
plt.yticks([y for y in range(min(y),max(y)+1,1)])
plt.show()
```
