```python
import pandas as pd #score 파일 불러오기 위한 패키지
import matplotlib.pyplot as plt
import matplotlib as mal

df=pd.read_csv('data/score.csv',index_col='지원번호')
df # x축 이름, y축 key(지원번호)
#한글설정
plt.rc('font',family='Malgun Gothic')
#사이즈값 설정
plt.rc('font',size=10) #각 축 값 글자 크기 지정(x,y축 값)
#한글깨지 방지
plt.rc('axes',unicode_minus=False) #xy축 -축의 한글 깨짐 방지(영문 사용 시 필요 없음)
```

```python
kor=list(df['국어'])
eng=list(df['영어'])
mat=list(df['수학'])
name=df['이름']

plt.Figure(figsize=(10,3))
plt.plot(name, kor, label='korea grade', color='red', marker='o', ls='--')
plt.plot(name, eng, label='english grade', color='green', marker='v', linestyle='-.')
plt.plot(name, mat, label='math grade', color='purple', marker='x')
plt.legend(loc=(1.02,0.5))
for idx, txt in enumerate(kor):
    plt.text(name[idx], kor[idx]+2, txt, color='red', ha='center', size=5)
plt.show()
```
