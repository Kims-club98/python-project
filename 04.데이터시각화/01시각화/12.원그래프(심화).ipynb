```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic') # 폰트 선택ㅇ
plt.rc('font',size=11) #폰트크기 11
plt.rc('axes',unicode_minus=False) #한글 깨짐 방지

import pandas as pd
```

```python
enroll=pd.read_csv('C:/Python/04.데이터시각화/01시각화/data/강좌.csv')
cou=pd.read_csv('C:/Python/04.데이터시각화/01시각화/data/수강.csv')
group=enroll.groupby('강좌번호').size().to_frame('인원수')
merge=cou.merge(group, on='강좌번호')
merge
```

```python
lables=merge['강좌번호']
values=merge['인원수']
colors = ['#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']
plt.pie(values,labels=lables,
        autopct='%.1%%',explode={0.05*len(values)},colors=colors)
```

```python
stu=pd.read_csv('data/학생.csv')
group=stu.groupby('학생학과').size()
group
lables=group.index
values=group.values
plt.pie(values, labels=lables
        ,autopct='%.0f%%')
plt.show()
```

```python
def grade(score):
    if score >=90:
        return 'A'
    elif score >=80:
        return 'B'
    elif score >=70:
        return 'C'
    elif score >=60:
        return 'D'
    else:
        return 'F'

```

```python
enroll=pd.read_csv('data/수강.csv')
enroll['학점']=enroll['점수'].apply(grade)
group=enroll.groupby('학점').size().to_frame()
```

```python
plt.pie(group.values, autopct='%.0f%%',
         labels=group.index + '학점',explode=[0.05]*len(group),wedgeprops={'width':0.8})
plt.show()
```
