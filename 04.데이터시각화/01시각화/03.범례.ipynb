```python
import matplotlib as mpl
import matplotlib.pyplot as plt

#한글설정
plt.rc('font',family='Malgun Gothic')
#사이즈값 설정
plt.rc('font',size=20)
#한글깨지 방지
plt.rc('axes',unicode_minus=False) #xy축 -축의 한글 깨짐 방지(영문 사용 시 필요 없음)
```

```python
year=[year for year in range(2020,2025, 1)] # range로 대체 가능 [2020,2021,2022,2023,2024]
height=[145,150,160,165,170]
plt.plot(year, height, label='년도별 신장') #x,y 값 지정(범례 지정 시 범례명 지정)
# plt.legend(loc='upper right') # 범례 생성(범례위치지정) (위: upper, 아래: bottom, 오른: right, 왼: left)
plt.legend(loc=(1,1)) # 범례위치 지정(외부로 보내기, 좌표 지정 느낌으로... (n,n))
plt.show()
```
