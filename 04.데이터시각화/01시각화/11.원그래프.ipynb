```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic') # 폰트 선택ㅇ
plt.rc('font',size=11) #폰트크기 11
plt.rc('axes',unicode_minus=False) #한글 깨짐 방지

import pandas as pd
df=pd.read_csv('data/score.csv')
```

```python
#원 그래프(lables, value 값이 필요)(전체 100으로 볼 때 몇 %가 필요한지?)
group=df.groupby('SW특기').size()
group #전체를 8로 보고 몇 명이 어떤 sw를 사용하는지 확인 할 때 '원 그래프'를 활용함

lables=group.index
values=group.values
lables, values
```

```python
plt.pie(values, labels=lables, 
        autopct='%.1f%%',startangle=90,counterclock=False) # 기본적인 퍼센트값 추출 autopct='%.1f%%' // startangle는 회전여부 결정(각도 기입)//counterclock은 T/F으로 결정하는데 좌우 반전 여부를 결정
plt.show()
```

```python
# explode 속성
values = [30, 25, 20, 13, 10, 2]
labels =['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC'] 

plt.pie(values,labels=labels, # 소수 둘째자리까지 출력: autopct='%.2f%%' 
        autopct='%.1f%%', startangle= 90, counterclock=False,
        explode=[0.05]*len(values)) # explode=[간격]*len(쪼갤 조각 수) 파이 조각을 나눠놓음]
plt.legend(loc=(1.1,0.5))
plt.show()
```

```python
def custom_autopct(pct):
    if pct <=2:
        return ''
    else:
        return f'{pct:.0f}%' #pct(퍼센트)가 2보다 작으면 공백, 그 외는 소수점 0번째자리 %로 출력함
```

```python
#wedgeprops 값
values = [30, 25, 20, 13, 10, 2]
labels =['Python', 'Java', 'Javascript', 'C#', 'C/C++', 'ETC'] 

plt.pie(values,labels=labels, # 소수 둘째자리까지 출력: autopct='%.2f%%' 
        autopct=custom_autopct, startangle= 90, counterclock=False,
        explode=[0.05]*len(values),  # explode=[간격]*len(쪼갤 조각 수) 파이 조각을 나눠놓음]
        wedgeprops={'width':0.5, 'edgecolor':'y', 'linewidth':0.5}, # wedgeprops는 도넛형태로 변경함을 의미(중괄호) width는 너비, edgecolor는 테두리 선색, linewith는 선 두께를 의미
         colors = ['#FFADAD','#FFD6A5','#FDFFB6','#CAFFBF','#9BF6FF','#A0C4FF']) 
plt.legend(loc=(1.1,0.5))
plt.show()
```
