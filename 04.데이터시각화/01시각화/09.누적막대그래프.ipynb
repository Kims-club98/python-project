```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic') # 폰트 선택ㅇ
plt.rc('font',size=11) #폰트크기 11
plt.rc('axes',unicode_minus=False) #한글 깨짐 방지

import pandas as pd
```

```python
#학생별 국어/영어/수학 점수의 꺾은선 그래프
df=pd.read_csv('data/score.csv')
plt.figure(figsize=(10,5))
plt.title('학생별 국어&영어&수학 점수', size=20)
plt.plot(df['이름'],df['국어'], label='Korean Grade',
          marker='o', ms=10, mec='r', mfc='y' ,mew=3) #plot 꺾은선 그래프
plt.plot(df['이름'],df['영어'], label='English Grade',
         lw=3, ls='--', c='brown')
plt.legend(loc=(1.1,0)) #범례위치, 범례는 꺾은선 그래프에서 설정하며 위치는 따로 지정함!!(그래프 외로 옮기고 싶으면 좌표로, 내부에서 편하게 설정하고 싶으면 lower/upper + right/left 조합을 작성) ex) loc='lower right' 
plt.ylim(0,110) # y축의 최대최솟값
yticks=[y for y in range(0,101,50)]#꺾은선 그래프에 '점' 등 포멧 형식을 구성하고 싶을 때 yticks를 새로 정의함
plt.yticks(yticks,[f'{y}점' for y in yticks]) # y축의 단위 값 (포멧형식 부여할 때)
xticks=df.loc[f'{df.loc[x,"이름"]}\n{df.loc[x, "지원번호"]}' for x in range(len(df))]
plt.xticks(df['이름'],xticks)
plt.xlabel('이름', size=15, loc='right', c='green') #x 레이블 지정하기(loc= right/left 기준)
plt.ylabel('점수',size=15,loc='top',c='g')# y 레이블 저장하기, (loc=top/bottom)

plt.show()
```

```python
# 학생별 이름, 키를 가지고 막대그래프 출력하기
plt.figure(figsize=(10,3))
plt.bar(df['이름'], df['키'],color='wheat', hatch='////////',width=0.8)
plt.ylim(160,210)
#문자열 열에 지원번호를 넣기
yticks=[y for y in range(160,211,10)]
plt.yticks(yticks,[f'{y}cm' for y in yticks])
plt.show()
```

```python
#학생별 평균 점수
df['평균']=df.apply(lambda row:row['국어':'사회'].mean(), axis=1)
plt.figure(figsize=(10,3))
plt.bar(df['이름'], df['평균'])
for idx, avg in enumerate(df['평균']):
    plt.text(idx, avg, f'{avg:.2f}점',ha='center')
plt.show()
```

```python
#SW특기별 몇 명이 있는가?(누적막대그래프)
group=df.groupby('SW특기').size() #sw특기는 index // 숫자는 values에서 가지고 옴
sw=group.index
size=group.values
plt.figure(figsize=(10,3))
plt.bar(sw, size)
```

```python
plt.bar(df['이름'], df['국어'],label='국어')
plt.bar(df['이름'], df['영어'],label='영어',bottom=df['국어']) #bottom 은 아래쪽에 무었을 두고 싶은가?
plt.bar(df['이름'], df['수학'],label='수학', bottom=df['국어']+df['영어']) # bottom은 국어/영어 밑에 있음 (결론, 국어 위에 영어 위에 수학을 둠=누적형태)
plt.ylim(0,300)
plt.legend(loc='upper left')
for idx in df.index:
    sum=df.loc[idx, '국어'] + df.loc[idx, '영어'] + df.loc[idx, '수학']
    plt.text(idx, sum+2, sum, ha='center')

plt.show()
```

```python
#사회 과학의 누적 막대 그래프
plt.bar(df['이름'],df['사회'],label='사회')
plt.bar(df['이름'],df['과학'],bottom=df['사회'],label='과학')
plt.legend(loc=(0,1))
for idx in df.index:
    sum=df.loc[idx, '사회'] + df.loc[idx, '과학']
    plt.text(idx, sum+2, sum, ha='center')
```

```python
#웹에 올릴 누적막대그래프(학생별 국어/영어/수학의 누적그래프)
df=pd.read_csv('data/score.csv')
plt.figure(figsize=(10,5))
plt.bar(df['이름'],df['국어'],label='국어')
plt.bar(df['이름'],df['영어'],label='영어',bottom=df['국어'])
plt.bar(df['이름'],df['수학'],label='수학',bottom=df['국어']+df['영어'],)
plt.legend()
plt.ylim(0,300)
plt.show()
```
