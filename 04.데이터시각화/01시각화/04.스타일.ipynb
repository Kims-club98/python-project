```python
import pandas as pd #score 파일 불러오기 위한 패키지
import matplotlib.pyplot as plt
import matplotlib as mal
```

```python
df=pd.read_csv('data/score.csv',index_col='지원번호')
df # x축 이름, y축 key(지원번호)
```

```python
# x, y축 설정하기
name= df['이름'] #x축
name #시리즈(list타입과 관계없이 seris도 사용 가능함)
height=df['키'] #y축
height
```

```python
#한글설정
plt.rc('font',family='Malgun Gothic')
#사이즈값 설정
plt.rc('font',size=10) #각 축 값 글자 크기 지정(x,y축 값)
#한글깨지 방지
plt.rc('axes',unicode_minus=False) #xy축 -축의 한글 깨짐 방지(영문 사용 시 필요 없음)
```

```python
plt.title('학생별 신장', size=20) #default값 10
plt.xlabel('이름') #축 이름
plt.ylabel('키(cm)')  #축 이름
plt.yticks([y for y in range(160,220,5)]) #y축 범위 160이상 220미만 5씩 증가
plt.plot(name, height, linewidth='2', marker='s') #x,y축 지정 + 라인의 굵기 지정(스타일, default =1),marker=는 각 데이터 구분 기호(o소문영어 원, v는 역삼각형,s는 사각형)
plt.show()
```

```python
plt.title('학생별 신장\n', size=20) #default값 10
plt.xlabel('이름') #축 이름
plt.ylabel('키(cm)')  #축 이름
plt.yticks([y for y in range(160,220,5)]) #y축 범위 160이상 220미만 5씩 증가
plt.plot(name, height, 
    linestyle=':', color='blue',#선 모양 변경(linestyle : 는 점선, Non은 선 없애기, '--'는 파선)
    linewidth='2', marker='s',markersize=10, markerfacecolor='yellow',markeredgecolor='magenta'
     , markeredgewidth=5) #마커 사이즈 변경(markersize), 마커 중심부 색깔 변경(markerfacecolor=), 마커 사이드 색깔 변경(markeredgecolor=), 마커 사이트 선 크기 변경(markeredgewidth=)



plt.show()
```

```python
#스타일은 축약어 사용 가능함
plt.title('학생별 신장\n', size=20, color='chocolate') #default값 10
plt.xlabel('이름') #축 이름
plt.ylabel('키(cm)')  #축 이름
plt.yticks([y for y in range(160,220,5)]) #y축 범위 160이상 220미만 5씩 증가
plt.plot(name, height,
        lw=5, c='darkgoldenrod', ls='--',
         marker='o', ms=10, mfc='yellow', mec='mediumspringgreen', mew=5 )

plt.show()
```

```python
#그래프의 사이즈 지정
plt.Figure(figsize=(10,3)) #그래프의 사이즈 지정
plt.ylim(160,220) #리미트값(시작값,끝값)을 지정해줌
plt.plot(name,height,marker='o')
#n번째 학생에 대한 데이터 출력하기(데이터 레이블)
for idx, txt in enumerate(height): #idx x축, txt y축
    plt.text(name[idx], height[idx]+2, f'{txt}cm',size=8, ha='center') #이름 기준으로 비교하고, 출력은 txt(키)
plt.show
```
