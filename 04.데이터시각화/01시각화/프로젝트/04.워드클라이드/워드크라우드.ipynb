```python
with open('영어가사.txt','r') as file:
    text=file.read()
print(text)
```

```python
from wordcloud import WordCloud #워드 클라우드 패키지 불러오기(앞 소문, 뒤 대문자)
import matplotlib.pyplot as plt
```

```python
wc=WordCloud().generate(text)
```

```python
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
import numpy as np
from PIL import Image

heart_mask=np.array(Image.open('heartmask.png'))
wc=WordCloud(
    width=800,
    height=200,
    background_color='wheat',
    min_font_size=20,
    mask=heart_mask).generate(text)


plt.figure(figsize=(15,15))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.savefig('goldenheart.png')
plt.show()
```
