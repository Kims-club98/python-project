```python
with open('영어가사(골든).txt','r') as file:
    text=file.read()
print(text)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
wc=WordCloud().generate(text)
plt. imshow(wc)
plt.axis('off')
plt.savefig('영어가사(골든).png')
plt.show()
```

```python
from konlpy.tag import Okt #한글 명사만 불러오는 패키지
from collections import Counter

with open('애국가.txt','r',encoding='utf-8') as file:
    text=file.read()
    print(text)
```

```python
okt=Okt()
nouns=okt.nouns(text) # text에서 명사만 추출해줌
nouns=[noun for noun in nouns if len(noun)>1] # nouns 에서 noun을 추출 / if를 통해 1글자 제외
nouns
```

```python
counter=Counter(nouns) # 각각의 noun이 몇 개 있는지 개수를 추출해주는 함수
counter
```

```python
wc=WordCloud(width=800, height=400,font_path='malgun.ttf').generate_from_frequencies(counter) #워드 크라우드 크기 설정  #빈도를 가지고 generate(생성)를 해줌
plt.imshow(wc)
plt.axis('off') # xy축의 단위를 숨기기
plt.show

```

```python
import numpy as np
from PIL import Image

heart_mask=np.array(Image.open('heartmask.png')) # 이미지 가지고 오기
wc=WordCloud(
    width=800, #너비
    height=200,#높이
    background_color='wheat', #배경색
    min_font_size=20, # 최소 글자 크기
    font_path='malgun.ttf', #글자 폰트(없으면 글자 깨짐)
    mask=heart_mask).generate(text)

plt.figure(figsize=(15,15))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.savefig('애국가heart.png')
plt.show()
```
