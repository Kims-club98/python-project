```python
data = {
    '영화' : ['명량', '극한 직업', '신과 함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번 방의 선물', '암살'],
    '개봉연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
    '관객수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270],
    '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
```

```python
import pandas as pd
df=pd.DataFrame(data)
#연도별 개봉연도에 따른 관객수의 합계(관객수가 많은 순으로 나열)(개봉연도는 1개만 있는 경우 대괄호 없어도 됨, 2개 이상일 경우는 [] 必)
group=df.groupby(['개봉연도'])['관객수'].sum().sort_values(ascending=False).to_frame()
df2=group.copy()
df2.reset_index(inplace=True)
df2.info() #개봉년도의 관객수가 0(Nan)인 경우 7개 O
df2['개봉연도']=df2['개봉연도'].astype(str) #columns 의 모든 데이터의 type을 변경해줌 그러면 Nan 데이터는 출력 X
df2.info()
```
