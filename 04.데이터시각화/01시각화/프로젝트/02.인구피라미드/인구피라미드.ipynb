```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic')
plt.rc('axes',unicode_minus=False)
import  warnings
warnings.filterwarnings('ignore')
```

```python
#2015년 남녀 인구 그래프 만들기

#남자 데이터 가져오기
import pandas as pd
df_m=pd.read_excel('data/2015년 08월 연령별인구현황.xlsx',skiprows=3, index_col='행정기관',usecols='B, E:Y') #skiprows는 열 만큼 가져오지 않겠음을 의미함, usecols은 엑셀에서 원하는 columns만 가지고 옴 E:Y는 E에서 Y까지 가지고 옴
df_m=df_m.head(1)
#data 전처리 과정(object를 int 로 변환), 필요한 정보만 추출
df_m.iloc[0]=df_m.iloc[0].str.replace(',','').astype(int) #df_m.iloc[0](칼럼 번호로 지정)
df_m.head(1)

#여자 데이터 가져오기
df_w=pd.read_excel('data/2015년 08월 연령별인구현황.xlsx',skiprows=3, index_col='행정기관',usecols='B, AB:AV') #skiprows는 열 만큼 가져오지 않겠음을 의미함, usecols은 엑셀에서 원하는 columns만 가지고 옴 
df_w.head(1)
#데이터 전처리과정
df_w.iloc[0]=df_w.iloc[0].str.replace(',','').astype(int)
df_w.columns=df_m.columns # 초기에는 m과 w의 칼럼이 다름(.1로 구분됨)(w=m을 통해 w의 colums를 동일하게 해줌)
df_w.columns

#남자 막대그래프
plt.figure(figsize=(7,5))
plt.barh(df_m.columns,-df_m.iloc[0]//1000) #//1000으로 하면 1000단위로 작성되며 df 앞 - 작성 시 바가 좌우반전됨
plt.barh(df_w.columns,df_w.iloc[0]//1000) #여자 그래프
plt.title('2015년 남녀인구분포', size=20)
plt.savefig('2015년 8월 남녀 인구.png')
plt.show
```
