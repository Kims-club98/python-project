```python
#산점도 그래프: 흩어저 있는 정도를 점으로 표현한 그래프
  #if 점이 모여 있다? 국어점수와 영어점수와 상관관계를 파악 가능함(국어를 잘하는 친구들이 영어를 잘아는가)
df=pd.read_csv('data/score.csv')
df['학년']=[3,3,2,1,1,2,2,3]
plt.scatter(df['국어'],df['영어'],
             sizes=df['학년']*500, # 크키를 통해 학년을 구분 가능함
             c=df['학년'], cmap='winter', # 색깔의 차이를 통해 학년 구분이 가능함 // cmap을 통해 색깔 테마를 지정해줄 수 있음
             alpha=0.3, #투명도 설정
             ) 
plt.xlabel('국어')
plt.ylabel('영어')
plt.colorbar(ticks=[1,2,3],orientation='horizontal',label='학년',shrink=0.5) #컬러 바 설정 #,orientation='horizontal'은 바의 세로위치(미기입), 가로위치(기입) 여부를 결정함 #shrink는 컬러바의 길이 여부를 결정함
for idx in range(len(df)): # 각 원의 이름을 설정
    x=df.loc[idx, '국어']#국어점수
    y=df.loc[idx, '영어']#영어점수
    name=df.loc[idx,'이름']
    plt.text(x,y,name, size=8, ha='center')
    
plt.show()
```

```python
#과학과 사회의 점수를 비교함
plt.scatter(df['사회'],df['과학'], sizes=df['학년']*500,
            c=df['학년'], cmap='summer')
plt.title('철산고 학생 사회/과학점수 분포')
plt.colorbar(ticks=[1,2,3],orientation='horizontal',label='철산고 학년',shrink=1)
for idx in range(len(df)):
    x= df.loc[idx, '사회']
    y= df.loc[idx, '과학']
    name=df.loc[idx, '이름']
    plt.text(x,y, name, ha='center', size=10)
```
