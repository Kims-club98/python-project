```python
import matplotlib.pyplot as plt
plt.rc('font',family='Malgun Gothic') # 폰트 선택ㅇ
plt.rc('font',size=11) #폰트크기 11
plt.rc('axes',unicode_minus=False) #한글 깨짐 방지

import pandas as pd
df=pd.read_csv('data/score.csv')
```

```python
#다중 막대그래프(국/영/수)
x=df.index*4 #1 사람이 4칸 차지(사람/국어/영어/수학을 차지)
plt.bar(x +0,df['국어'],label='국어',width=1) # x가 3
plt.bar(x+1,df['영어'],label='영어',width=1)# x가 4
plt.bar(x+2,df['수학'],label='수학',width=1)# x가 5 #bar 가 1개일때 width 는 1 이후 개수가 늘 수록 0.5,
xticks=[idx+1 for idx in x] #x축 중심 맞추기 위한 식
plt.xticks(xticks,df['이름']) # 중심을 맞추고 index 번호를 df['이름']으로 변경
for idx in df.index:
    x=idx*4
    plt.text(x,df.loc[idx, '국어'], df.loc[idx, '국어'], ha='center', size=8 ) #plt.text(텍스트에 출력되는 x, 텍스트에 출력되는 y, 실제로 출력되는 값)
    plt.text(x+1,df.loc[idx, '영어'], df.loc[idx, '영어'], ha='center', size=8 )
    plt.text(x+2,df.loc[idx, '수학'], df.loc[idx, '수학'], ha='center', size=8 )
plt.legend()
plt.show()
```

```python
#사회와 과학점수 출력
x=df.index*3
plt.bar(x+0,df['사회'],label='사회',width=1,color='red')
plt.bar(x+1,df['과학'],label='과학',width=1,color='blue')
xticks=[idx+1 for idx in x]
plt.xticks(xticks,df['이름'])
for idx in df.index:
    x=idx*3 ## idx 기준으로 하기
    plt.text(x,df.loc[idx, '사회']+1, df.loc[idx, '사회'], ha='center', size=8 )
    plt.text(x+1,df.loc[idx, '과학']+1, df.loc[idx, '과학'], ha='center', size=8 )
plt.legend(loc=(1,1))
```

```python
# 학생별 국어,영어,수학 다중막대그래프
df=pd.read_csv('data/score.csv')
x=df.index*4
plt.bar(x,df['국어'],width=1,label='국어') # x축이 0,4,8,16,32...칸마다 기입됨
plt.bar(x+1,df['영어'],width=1,label='영어')
plt.bar(x+2,df['수학'],width=1,label='수학')
xticks =[xticks+1 for xticks in x]
plt.xticks(xticks, df['이름']) #x축의 x값과 이름 지정
plt.legend(loc=(1.1,0))
plt.show()
```
