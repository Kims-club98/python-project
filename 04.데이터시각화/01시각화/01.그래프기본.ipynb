```python
plt. title('꺾은선 그래프', size=20) #그래프 제목, 크기
plt.plot=(x,y)
plt.xlabel('x축', color='red', size=15, loc='right') #x축 제목, 색깔, 사이즈 조절, 위치조절(left, right)
plt.ylabel('y축', color='blue', size=13, loc='top')#y축 제목, 색깔, 사이즈 조절, 위치 조절(top, bottom)
plt.show()
```
