```python
#기본 세팅
import pandas as pd #score 파일 불러오기 위한 패키지
import matplotlib.pyplot as plt
import matplotlib as mal

#한글설정
plt.rc('font',family='Malgun Gothic') #글꼴은 변경 가능
#사이즈값 설정
plt.rc('font',size=10) #각 축 값 글자 크기 지정(x,y축 값)
#한글깨지 방지
plt.rc('axes',unicode_minus=False) #xy축 -축의 한글 깨짐 방지(영문 사용 시 필요 없음)
```

```python
#파일 가져오기
df=pd.read_csv('data/score.csv',index_col='지원번호')
df # x축 이름, y축 key(지원번호)
name=df['이름']
height=df['키']
```

```python
plt.barh(name, height) #세로 막대형을 가로 막대형으로 변경함(x레이블이 y레이블로 이동/y레이블이 x레이블로 옮겨짐) "행렬전환"
plt.xlabel('학생키') #기존 name 위치에서 plt.barh로 xy가 전환됨
plt.xlim(160,210)
plt.show
```

```python
plt.Figure(figsize=(10,3)) #그래프 size 변경해줌
bar=plt.bar(name, height, hatch='x',color='none', ec='g',) #그래프 바 안에 문양을 넣어줌, color 색 없애기 'none'
plt.ylim(160,210)
bar[1].set_hatch('/') # 각각의 hatch 스타일 변경 필요시 위 내역을 bar변수 선언하고 하나씩 넣어주면 됨
bar[2].set_hatch('o')
bar[3].set_hatch('')
for idx, h in enumerate(height):
    plt.text(idx, h+1, h, ha='center') #가운데 정렬 #h+1은 y좌표를 약간 올림
plt.show()
```

```python
df['평균']=df.apply(lambda row:row['국어':'사회'].sum()/5,axis=1)
df['평균']
```

```python
#학생 별 국어 점수

plt.bar(name, df['평균'])
plt.title('학생별 평균')
plt.Figure(figsize=(10,5))
plt.ylim(0,100)
plt.yticks([y for y in range(0,101,5)])
for idx, d in enumerate(df['평균']):
    plt.text(name[idx], df['평균'][idx]+1, d,size=8, ha='center')

plt.xlabel('학생이름')
plt.ylabel('평균점수')
plt.tight_layout()
plt.show()
```

```python
#학생별 평균
#1. ylim(0,100)
#2. ytick(0,5,10...) 0부터 100까지
#3. xlabel= 학생이름
#4. ylabel= 평균점수
#5. 학생별 막대별 평균점수
#6. 그래프 제목, 학생별 평균
#7. 그래프 사이즈는 10,5
```
