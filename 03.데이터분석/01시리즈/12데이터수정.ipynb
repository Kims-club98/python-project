```python
import pandas as pd
```

```python
df=pd.read_csv('data/score.csv',index_col='지원번호')
df
```

```python
# 북산고 - 산북고 /능남고 - 남능교 수정하기
df2=df.copy() #df2로 카피
filt=df2['학교']=='북산고' #북산고를 찾음
df2.loc[filt,'학교']='산북고' #df2에서 북산고를 산북고로 바꿈
filt=df2['학교']=='능남고'
df2.loc[filt,'학교']='남능교'
df2
```

```python
df3=df.copy()
df3['학교'].replace({'북산고':'산북고','능남고':'남능고'},inplace=True)
df3
```

```python
df4=df.copy()
df4['SW특기']=df['SW특기'].str.capitalize()
df4['학교']=df4['학교'] + '등학교'
df4['총점']=df4['국어']+df4['영어']+df4['수학']+df4['과학']+df4['사회']
df4['평균']=df4['총점']/5
df4['결과']='Fail'
#평균이 70점 이상은 Pass 출력
filt=df4['평균'] >= 70
df4.loc[filt,'결과']='Pass'
df4
```

```python
df5=df.copy()
df5.drop(index='4번') # 4번 내역(서태웅) 삭제
df5.drop(columns=['키','SW특기']) #키와 SW특기 삭제
```
