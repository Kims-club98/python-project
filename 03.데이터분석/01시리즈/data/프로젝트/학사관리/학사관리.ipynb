```python
#학생 csv를 이용해 수강신청한 강좌 수 추출 tt
def stu_list():
    강좌=pd.read_csv('강좌.csv')
    수강=pd.read_csv('수강.csv')
    학생=pd.read_csv('학생.csv')
    merge=pd.merge(강좌,수강,how='left',on='강좌번호')
    merge2=pd.merge(학생,merge,how='left',on='학생번호')
    group=merge2.groupby(['학생번호','학생이름','학생학과','학년','생년월일','지도교수']).size().to_frame('수강 강좌수')
    group.reset_index(inplace=True)
    return group
```

```python
#tt 결과
def stu_list():
    enroll = pd.read_csv('수강.csv')
    group = enroll.groupby('학생번호').size().to_frame('강좌수')
    stu = pd.read_csv('학생.csv')
    merge = stu.merge(group, how='left', on='학생번호')
    merge.fillna({'강좌수':0}, inplace=True)
    merge.set_index('학생번호', inplace=True)
    merge['강좌수']=merge['강좌수'].astype(int)
    return merge

stu_list()
```
