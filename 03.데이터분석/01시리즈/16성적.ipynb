```python
#지원번호를 입력받으면 관련하여 정보를 제공해줌
no=input('지원번호>')
filt=score['지원번호']==int(no) #score의 지원번호는 지원번호와 같음
idxs=score.index[filt] 
if len(idxs)==0:
    print('검색결과 없음')
else:
    row=score.loc[idxs[0]]
    for col in cols:
        print(f"{col}:{row[col]}")
```

```python
no=8
filt=score['지원번호']==no #no는 score의 지원번호임
idx=score[filt].index #row는 지원번호(filt)에 관한 내역
if len(idx)==0:
    "No Data"
else:
    row=score.loc[idx[0]]
    for col in cols:
        print(f"{col}:{row[col]}")
    score.drop(index=idx[0],inplace=True)
```
