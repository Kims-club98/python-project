```python
def change_no(no): #지원번호의 '번'을 제거+ no를 정수로 변경(원래 문자열)
    no= no.replace('번','')
    no=int(no)
    no=f'{no:02d}'
    return no
```

```python
# change_no의 확인검증 
no=change_no('10번')
no, type(no)
```

```python
df2=score.copy()
df2.reset_index(inplace=True)
df2['지원번호']=df2['지원번호'].apply(change_no) #1번 을 01 int로 변경을 적용함
df2.set_index('지원번호', inplace=True) #인덱스를 새로 지정(reset.index 원래대로 인덱스 돌리기)
df2
```
