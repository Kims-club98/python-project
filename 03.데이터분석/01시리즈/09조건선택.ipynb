```python
import pandas as pd
```

```python
df=pd.read_csv('data/학교키순정렬.csv')
df.rename(columns={'Unnamed: 0':'지원번호'}, inplace=True)
df.set_index('지원번호',inplace=True)
print(df) #unnamde: 0을 '지원번호'로 이름을 변경함
```

```python
df.sort_index(inplace=True)
df
```

```python
filt=df['키']>=185 #참인 경우에만 True를 출력함
filt
```

```python
# 방법2
df[df['키']>=180]
```

```python
#조건에 만족 x(+보고싶은 것만)
df[~filt][['이름','키']]
```

```python
df.loc[~filt,'이름':'키']
```

```python
#북산고 면서, 키가 185 이상
filt=(df['키']>=185)&(df['학교']=='북산고')
df[filt][['키','학교']]
```

```python
#키가 170보다 작거나, 키가 200보다 큰 학생들 (or= |)
filt=(df['키']<170) | (df['키']>200)
df[filt]
```

```python
#'송'으로 시작하는 이름만 적용
filt=df['이름'].str.startswith('송')
df[filt]
```

```python
#'태'가 포함된 것만 출력
filt=df['이름'].str.contains('태')
df[filt]
```

```python
# 이름, 학교만 보고 싶음
df[~filt][['이름','학교']]
```

```python
#sw를 Javascript만 출력하고 싶음
langs=['PYTHON','Java','C']
filt=df['SW특기'].str.capitalize().isin(langs)
df[filt]
```

```python
#Java가 들어가거나, Nan값을 처리하는 것
filt=df['SW특기'].str.contains('Java',na=False) #na 값을 False로 처리함(오류 방지)
df[filt]
```
