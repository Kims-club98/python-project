```python
import pandas as pd
score = pd.read_csv('data/score.csv', index_col='지원번호')
score
```

```python
df=score.copy()
df.loc['9번','이름']='홍길동' #loc: 더함의 이름 9번행의 '이름'열에 '홍길동'을 추가함을 의미
df
```

```python
df.reset_index(inplace=True) # df DF 초기화
df
```

```python
#0번에서 00 형식으로 변경해주는 함수 def change_no(no)
def change_no(no): #선언
    no=no.replace('번','')  # '번'을 공백으로 변경
    no=int(no) #str인 no int형식 전환(정수)
    no=f'{no:02d}' # no 표기형식 변경 00형태
    return no #반환
```

```python
# change_no 적용
df2=score.copy()
df2.reset_index(inplace=True) #df2 형식 초기화
df2['지원번호']=df2['지원번호'].apply(change_no) # df2 지원형식에 change_no 적용
df2.set_index('지원번호',inplace=True) # 지원번호를 index열제목 설정
df2
```

```python
#홍길동 행 추가
df2.loc['09','이름']='홍길동'
df2# 이름만 추가했으므로, 나머지는 NaN 처리됨
```

```python
#인덱스 번호 자동 지정
max=df2.index.max()
new_no=str(int(max)+1) #최초 max는 문자열 '09', but +1을 위해 int로 정수화 후 str로 다시 문자화함
new_no
```

```python
# new_no 적용
df2.loc[new_no,'이름']='강감찬'
df2
```

```python
#일부 열만 추출
df_info=score[['이름','학교','키','SW특기']] # :(범위지정) [] // ,(각각지정) [[]]
df_info
```

```python
info=df_info.copy()
info.reset_index(inplace=True)
info
```

```python
# change_no 적용
info['지원번호']=info['지원번호'].apply(change_no)
info.set_index('지원번호',inplace=True) # 지원번호를 index열제목으로 설정
info
```

```python
#목록출력
cols=info.columns #['이름', '학교', '키', 'SW특기'] 열 제목
index=info.index #'01', '02', '03', '04', '05', '06', '07', '08' 행 제목
for idx in info.index: # info의 행 제목에서 선택함
    row=info.loc[idx] # row는 선택된 행 제목에서 더함
    print(f'지원번호:{idx}', end=' ') #' ' 로 각 구분
    for col in cols:# 열 제목에서 선택
        print(f"{col}:{row[col]}",end=' ')# row[col]은 열 제목에서 행제목과 부합하는 것 즉 "값"
    print() # idx, col 추출 이후 구분목적 for col문 밖에 위치함
    print('-'*50)

        
```
