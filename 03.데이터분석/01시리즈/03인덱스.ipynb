```python
import pandas as pd 
```

```python
data = {
    '이름' : ['채치수', '정대만', '송태섭', '서태웅', '강백호', '변덕규', '황태산', '윤대협'],
    '학교' : ['북산고', '북산고', '북산고', '북산고', '북산고', '능남고', '능남고', '능남고'],
    '키' : [197, 184, 168, 187, 188, 202, 188, 190],
    '국어' : [90, 40, 80, 40, 15, 80, 55, 100],
    '영어' : [85, 35, 75, 60, 20, 100, 65, 85],
    '수학' : [100, 50, 70, 70, 10, 95, 45, 90],
    '과학' : [95, 55, 80, 75, 35, 85, 40, 95],
    '사회' : [85, 25, 75, 80, 10, 80, 35, 95],
    'SW특기' : ['Python', 'Java', 'Javascript', '', '', 'C', 'PYTHON', 'C#']
}
```

```python
df=pd.DataFrame(data, index=['1번','2번','3번','4번','5번','6번','7번','8번'])
df
```

```python
df.index #df3의 인덱스만 확인
```

```python
df3=pd.DataFrame(data,index=['1번','2번','3번','4번','5번','6번','7번','8번'])
df3 #데이터를 dataframe으로 변경하고 인덱스도 변경함
```

```python
df3.index.name='지원번호'
df3
```

```python
df3.reset_index() #반영은 X, 확인용도
df3
```

```python
df3.reset_index() #index번호가 일반칼럼으로 이동
```

```python
df3.reset_index(drop=True) #인덱스 네임이 아예 사라짐(적용되지 않음)
df3
```

```python
df3.reset_index(inplace=True) # 인덱스 네임 적용
df3
```

```python
df.set_index('이름') #확인용, 적용 X
df
```

```python
df.set_index('이름',inplace=True) #이름 내역 인덱스 적용
df
```

```python
df[['학교','키']]['채치수':'서태웅'] # 학교/키만 채치수 이후의 사람을 확인
# 적용목적: inplcea=true 만 가능
```

```python
df.sort_index() #index 기준 오름차순 세팅
```

```python
df.sort_index(ascending=False) #index 기준 내림차순 세팅
```

```python
df.sort_index(ascending=True) #오름차순 정렬 적용은 X
df
```

```python
df.sort_index(ascending=False, inplace=True)
df # 실제 적용 inplace
```

```python
df.sort_values(['학교','키'], ascending=False) #학교에서 작은경우, 같은 경우 키가 작은 경우 추출
```
