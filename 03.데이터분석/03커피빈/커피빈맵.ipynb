```python
#csv 주소 토대로 좌표 받기(API)
def get_location(address):
    url="https://dapi.kakao.com/v2/local/search/address.json?query="+address #주소를 기반으로 좌표를 받기
    headers={"Authorization": 'KakaoAK e779cdef1331864a0977eaff9b2fed22'}
    res=requests.get(url,headers=headers).json()
    if res['documents']:
        x=res['documents'][0]['x']
        y=res['documents'][0]['y']
    else:
        x=np.nan
        y=np.nan
    return x, y #자료 받아오고 return 해줘야 함
    
    

x, y=get_location('서울시')
x, y
```

```python
store=pd.read_csv('커피빈매장.csv')
store['X']=''
store['Y']=''
store.head(2)
```

```python
for i in store.index:
    address=store.loc[i,'Address']
    x, y = get_location(address)
    print(x, y)
    store.loc[i,'X']=x
    store.loc[i,'Y']=y
```

```python
store.isnull().sum()
store.dropna(inplace=True)
store.to_csv('커피빈맵.csv',index=False)
```

```python
address='서울시 강남구'
filt=store['Address'].str.contains(address)
df= store[filt]
store=pd.read_csv('커피빈맵.csv',index_col=False)
```

```python
import folium #지도 만드는 함수(진짜 맵)
import warnings



def get_map(address):
    filt=store['Address'].str.contains(address)
    df=store[filt]
    if len(df.index)==0:
        print('없음')
    else:
        first=df.head(1) #첫번째 데이터 가져오기
        x=first['X']
        y=first['Y']
        map=folium.Map(location=[y,x], zoom_start=15) #맵을 형성하며, 위치는 y x로 초기 크기는 15로 지정 
        for idx, row in df.iterrows():
            head=row['Name']
            x=row['X']
            y=row['Y']
            folium.Marker(location=[y,x],popup=folium.Popup(f'{idx},{head}', max_width=200), icon=folium.Icon(icon_color='purple',icon='triangle')).add_to(map)
        map.save(f'{address}.html')       
        print('있음')
get_map('서울')
```
